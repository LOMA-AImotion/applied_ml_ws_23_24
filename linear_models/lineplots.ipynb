```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Set the seed for reproducibility
np.random.seed(42)

# Generate random parameters for 10 lines
num_lines = 10
w0_values = np.random.uniform(-10, 10, num_lines)
w1_values = np.random.uniform(-5, 5, num_lines)

# Generate x values
x_values = np.linspace(0, 10, 100)

# Generate y values for each line
y_values = np.array([w0 + w1 * x_values for w0, w1 in zip(w0_values, w1_values)])

# Set font size
font_size = 16

# Create a Seaborn plot with two subplots
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 6))

# Plot 10 random lines on the left subplot
axes[0].set_title('Data Space', fontsize=font_size)
axes[0].set_xlabel('X', fontsize=font_size)
axes[0].set_ylabel('Y', fontsize=font_size)
for i, y in enumerate(y_values):
    line_color = sns.color_palette('tab10', num_lines)[i]
    sns.lineplot(x=x_values, y=y, ax=axes[0], color=line_color, label=f'Line {i+1}')

# Scatter plot with the same colors on the right subplot
sc = axes[1].scatter(w0_values, w1_values, c=sns.color_palette('tab10', num_lines), alpha=0.8)
axes[1].set_title('Parameter Space', fontsize=font_size)
axes[1].set_xlabel('w0', fontsize=font_size)
axes[1].set_ylabel('w1', fontsize=font_size)

# Increase tick label font size
for ax in axes:
    ax.tick_params(axis='both', which='major', labelsize=font_size)

# Adjust layout to prevent overlapping
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# Define the dataset X and target values y
X = np.array([[1, 1], 
              [1, 2.5], 
              [1, 3], 
              [1, 3.4], 
              [1, 4]])

y = np.array([1, 1.4, 2.5, 1.9, 3.2])

# Hyperparameters
learning_rate = 0.01
num_epochs = 100

# Initialize weights randomly
w = np.random.rand(2)

# Function to compute the mean squared error
def mse(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)

# Gradient descent optimization
def gradient_descent(X, y, w, learning_rate, num_epochs):
    weights_history = [w.copy()]
    mse_history = [mse(y, X @ w)]

    for epoch in range(num_epochs):
        # Calculate the predicted values
        y_pred = X @ w

        # Calculate the gradient
        gradient = -2 * X.T @ (y - y_pred) / len(y)

        # Update weights
        w -= learning_rate * gradient

        # Save weights and MSE for visualization
        weights_history.append(w.copy())
        mse_history.append(mse(y, X @ w))

    return weights_history, mse_history

# Run gradient descent
weights_history, mse_history = gradient_descent(X, y, w, learning_rate, num_epochs)

# Animation function
def update(frame):
    plt.clf()
    plt.scatter(X[:, 1], y, label='Data points')
    plt.plot(X[:, 1], X @ weights_history[frame], color='red', label=f'Epoch {frame}/{num_epochs}')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Linear Regression Learning Process')
    plt.legend()

# Create animation
fig, ax = plt.subplots(figsize=(8, 6))
animation = FuncAnimation(fig, update, frames=num_epochs, interval=100, repeat=False)

# Save the animation as a video file (requires ffmpeg)
animation.save('linear_regression_learning.mp4', writer='ffmpeg', fps=10)

# Display the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# Define the dataset X and target values y
X = np.array([[1, 1], 
              [1, 2.5], 
              [1, 3], 
              [1, 3.4], 
              [1, 4]])

y = np.array([1, 1.4, 2.5, 1.9, 3.2])

# Hyperparameters
learning_rate = 0.01
num_epochs = 100

# Initialize weights randomly
w = np.random.rand(2)

# Function to compute the mean squared error
def mse(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)

# Gradient descent optimization
def gradient_descent(X, y, w, learning_rate, num_epochs):
    weights_history = [w.copy()]
    mse_history = [mse(y, X @ w)]

    for epoch in range(num_epochs):
        # Calculate the predicted values
        y_pred = X @ w

        # Calculate the gradient
        gradient = -2 * X.T @ (y - y_pred) / len(y)

        # Update weights
        w -= learning_rate * gradient

        # Save weights and MSE for visualization
        weights_history.append(w.copy())
        mse_history.append(mse(y, X @ w))

    return weights_history, mse_history

# Run gradient descent
weights_history, mse_history = gradient_descent(X, y, w, learning_rate, num_epochs)

# Animation function
def update(frame):
    plt.clf()

    # Plot linear regression on the left subplot
    plt.subplot(1, 2, 1)
    plt.scatter(X[:, 1], y, label='Data points')
    plt.plot(X[:, 1], X @ weights_history[frame], color='orange', label=f'Epoch {frame}/{num_epochs}')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Linear Regression Learning Process')
    plt.legend()

    # Plot loss surface on the right subplot (contour plot)
    plt.subplot(1, 2, 2)
    w0_vals = np.linspace(-1, 3, 100)
    w1_vals = np.linspace(-1, 3, 100)
    W0, W1 = np.meshgrid(w0_vals, w1_vals)
    loss_surface = np.array([[np.sqrt(mse(y, X @ np.array([w0, w1]))) for w1 in w1_vals] for w0 in w0_vals])
    contours = plt.contour(W0, W1, loss_surface, cmap='viridis', levels=20, alpha=0.8)
    plt.scatter(np.array(weights_history[:frame + 1])[:, 0], np.array(weights_history[:frame + 1])[:, 1],
            color='red', s=30, label=f'Epoch {frame}/{num_epochs}')
    plt.xlabel('w0')
    plt.ylabel('w1')
    plt.title('Root Mean Squared Error Loss Surface')
    plt.legend()

# Create animation
fig = plt.figure(figsize=(15, 6))
animation = FuncAnimation(fig, update, frames=num_epochs, interval=100, repeat=False)

# Save the animation as a video file (requires ffmpeg)
animation.save('linear_regression_learning_with_contour_rmse_loss_surface.mp4', writer='ffmpeg', fps=10)

# Display the plot
plt.show()
```
